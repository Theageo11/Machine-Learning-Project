<font size='6'> Linear Regression

```python
data=data.drop(columns=['90% Confidence Level - Low','90% Confidence Level - High','95% Confidence Level - Low','95% Confidence Level - High'])
data.columns
```

```python
data.drop(columns=['Year','Indicator','BCHC Requested Methodology','Source','Indicator Category','Methods','Notes'], inplace=True)
data.columns
```

```python
data = data.dropna(subset=['Value'])
data.info()
```

```python
df_categorical_col = data.select_dtypes(exclude=np.number).columns
df_categorical_col
```

```python
df_numeric_col = data.select_dtypes(include=np.number).columns
df_numeric_col
```

```python
df_onehot = pd.get_dummies(data[df_categorical_col])
df_onehot.head(5)
```

```python
df_onehot.info()
```

```python
df_after_encoding = pd.concat([data[df_numeric_col],df_onehot], axis = 1)
df_after_encoding.head(5)
df_after_encoding.dropna()
```

```python
df_after_encoding.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import linear_model

y=df_after_encoding['Value'].values
X=df_after_encoding.drop(columns='Value').values
X.reshape(-1,1)
```

```python
X=np.asanyarray(X)
X.shape
```

```python
y=np.asanyarray(y)
```

```python
y.shape
```

```python
train_x, test_x, train_y, test_y = train_test_split(X,y,test_size=0.3,random_state=2)
#train_x=train_x.reshape(-1,1)
train_y=train_y.reshape(-1,1)
#test_x=test_x.reshape(-1,1)
test_y.reshape(-1,1)
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
linf=linear_model.LinearRegression()
linf
```

```python
linf.fit(train_x,train_y)
```

```python
predlinf=linf.predict(test_x)
predlinf
```

```python
linf.coef_
```

```python
linf.intercept_
```

```python
from sklearn import metrics
test_aberror=metrics.mean_squared_error(test_y,predlinf)
rmse=np.sqrt(test_aberror)
print('linear regression Rmse is',rmse)
```

<font size='6'> KNN Model

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=10)
knn
```

```python
knn.fit(train_x,train_y)
```

```python
predknn=knn.predict(test_x)
predknn
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
rmse_dict = {}
rmse_list = []
for k in range(1,11):
    knn = KNeighborsRegressor(n_neighbors = k).fit(train_x,train_y)
    Y_predict = knn.predict(test_x)
    rmse = sqrt(mean_squared_error(test_y,Y_predict))
    rmse_dict.update({k:rmse})
    rmse_list.append(rmse)
    print("Rmse for k = {} is {}" .format(k,rmse))
```

<font size='5'> Comparing the two models' Rmse, Linear model has a smaller Rmse

<font size='5'> Regression Tree also has the close Rmse

<font size='9'> Kmeans Clustering

```python
## Standardlize X_km
X_km=df_after_encoding[df_after_encoding.columns]
from sklearn.preprocessing import StandardScaler
Clus_dataSet = StandardScaler().fit_transform(X_km)
Clus_dataSet
```

```python
from sklearn.cluster import KMeans
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 2)
k_means.fit(X_km)
labels = k_means.labels_
print(labels)
```

```python
df_after_encoding['labels']=labels
df_after_encoding.head(12)
```

```python
df_after_encoding.groupby('labels').mean()
```
